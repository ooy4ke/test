##discord.pyを動かすには
ここからは簡単なdiscordのbotの作り方について書いていきます。本を参考にして行っていますが、本の通りに行ってもうまくいかない部分がかなりあったので、自分の対処法も書いていきます。<br><br><br>

1. CLIを開く  

今回、ライブラリーのインストールやプログラムの実行は、OS上のCLIから行います。Linuxならターミナル、Windowsならコマンドラインプロンプトを開いてください。Windowsでコマンドラインプロンプトを開くには、windowsメニューを開いてcmdと入力し、エンターキーを押します。

2. discord.pyをインストールする  

Bot開発を支援する イブラリー「discord.py」をインストールするのにCLIに下を入力します。

```
python 3 -m pip install -U discord.py ← Linux の 場合  
py -3 -m pip install -U discord.py ← Windows の 場合
```

自分はここでエラーが出たのでそれについて書いていきます。エラーが出ない人は飛ばして読んでください。


- pipに関するエラー  
pipが最新ではないときエラーが出るようで、エラーのところにpipをアップデートする方法がかかれるのでそれをコピーして入力するとアップデートできます。  
pipがファイルの中に見つからないというエラーが出たときは`pip -V`と入力し、pipのあるフォルダを調べて、pipのファイルをユーザーファイルに持ってくるとうまくいきました。

- `Could not build wheels for aiohttp, which is required to install pyproject.toml-based projects`というエラー

![pythonダウンロード](https://drive.google.com/uc?export=view&id=1VLI7bsEl8qrnzhe2X7vJ_ooRJe-RIvuu)



3. discord.pyのインストールがうまくいったか確認する  

インストールがうまくいったか確認するためにコマンドプロンプトに下を入力します。数行のパッケージ情報が表示されればインストールはうまくいっています。

```
python 3 -m pip show discord.py← Linuxの場合   
py -3 -m pip show discord.py ← Windowsの場合
```

4. botをdiscordに接続する  

先ほどインストールしたVS Codeに下の内容を入力し、○○○○のところに先ほどコピーしたトークンを入力します。これを「bot.py」と名前を付けて保存します。

```
import discord
from discord.ext import commands

intents = discord.Intents.default()
intents.message_content = True  

TOKEN = "○○○○○○○○○○○○○○○○○○○○"  
bot = commands.Bot(command_prefix="?",intents=intents)
  
bot.run(TOKEN)
```


CLIに下を入力して、discordのbotがオンラインになっていたら成功です。

```
python3 bot.py← Linuxの場合
py -3 bot.py← Windowsの場合
```


5. イベントとは  

ここまででbotとしてdiscordにログインする処理が実装できました。ここからはイベントに反応して処理するようコードを変更します。イベントとはdiscord上で発生する様々な出来事です。


- botが準備完了したとき
- チャンネルにメッセージが投稿されたときなど<br><br>



その他のイベントについてはdiscord.pyの公式ドキュメントに掲載されています。それぞれのイベントについて、イベントごと呼ばれるイベントリスナーのメソッド名が決まっているので、そのメソッドで処理内容を定義します。<br><br><br>





6. イベントを実行する  

ここからは「 Botが準備完了したとき」のイベントを拾って、ログインしたBotアカウントのユーザー名をCLI上に表示する処理を実装します。  
先ほど書いたコードのbot. run(TOKEN)の直前に、下の内容を追加します。


```
@bot.event

async def on_ready():
   print(f"ログイン完了 ユーザー名: {bot.user.name}")
```


先ほどと同じく下を入力して

```
python3 bot.py← Linuxの場合
py -3 bot.py← Windowsの場合
```

CLIに`ログイン完了 ユーザー名:<Botアカウントのユーザー名> `と出力されたら成功です。

7. botのコマンドを定義する  

ユーザーからの「コマンド」を受け取って処理を行うようにコードを変更します。  
botにコマンドを実装したいときは、コマンド名をそのままメソッド名に採用し、@bot.command()デコレータをつけたメソッドを定義すれば完了です。<br><br>



下のコードを先ほどと同じく`bot. run(TOKEN)`の前に追加します。

```
@bot.command()

async def hello(ctx):
   await ctx.send(f"hello, {ctx.author.name}.")

```

discordを開いて`?hello`と入力してみて、botが`hello ＜ユーザー名＞`と返してくれれば完成です。<br><br>
  
今回は簡単な内容をしましたが、もっと多様な処理を入れたいときは「discord.py公式ドキュメント」と「RoboDanny」を参考にしてみてください。
