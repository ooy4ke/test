## discordが持っている情報
- ユーザーに関する情報
  - パスワード
  - メールアドレス
  - 名前
  - 入っているサーバー
  - フレンド
  - アイコン  
- サーバーに関する情報
 - 名前
 - 属しているユーザーとそのユーザーに関する情報（名前、アイコン、ロールなど）
 - アイコン
 - チャンネルに関する情報
 - 入っているbot  
 - チャットの履歴













## discordの画面
### サーバーなどの選択
サーバーを選択したり、追加したりする機能が下の写真の位置にある。  
➀ダイレクトメッセージを選択する。  
②サーバーを選択する。  
③サーバーを追加する。  
④発見対処のサーバーを見つける。

### チャンネルの選択など
ボイスチャンネルやテキストチャンネルの選択などの機能が写真の位置にある。  
➀サーバー名が表示されており、サーバーの設定などを行う。  
②イベント作成などを行う。  
③テキストチャンネルの追加、選択、編集、招待を作製を行う。  
④ボイスチャンネルの追加、選択、編集、招待を作製を行う。
